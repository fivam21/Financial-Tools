```python
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# SIMULATED WIENER PROCESS
simulations = []
for n in range(5):
    results = []
    for t in range(1000):
        dist = np.random.normal(0, math.sqrt(0.1))
        results.append(dist)
    csum = np.cumsum(results)
    simulations.append(csum)

plt.figure(figsize=(5,5))
plt.xlabel("Time")
plt.ylabel("Value")
plt.title("Simulated Wiener Processes")
for i in range(len(simulations)):
    plt.plot(simulations[i])
plt.show()
```

```python
# ABSOLUTE QUADRATIC VARIATION
T = 1
N = [10,100,1000, 10000]
simulations = []
for i in N:
    dt = T/i

    # Generate 1000 simulations of the Wiener process
    dW = np.sqrt(dt) * np.random.normal(size=(i,1000))
    W = np.cumsum(dW, axis=0)

    # Compute the quadratic variation of each simulation
    quad_var = np.sum(np.abs(dW), axis=0)
    df = pd.DataFrame(quad_var)
    simulations.append(df)

    # Compute the maximum and minimum values of the quadratic variation 
    max_quad_var = np.max(quad_var)
    min_quad_var = np.min(quad_var)

    # Check that the maximum and minimum values are finite
    print(f"Maximum absolute quadratic variation for dt = 1/{i}: {max_quad_var:.3f}")
    print(f"Minimum absolute quadratic variation for dt = 1/{i}: {min_quad_var:.3f}\n")

dims = int(len(N)/2)
plt.figure(figsize=(8, 8))
plt.subplots_adjust(hspace=0.5)
plt.suptitle("Absolute Quadratic Variation", fontsize=18, y=0.95)

for n in range(1,len(simulations)+1):
    ax = plt.subplot(dims, dims, n)
    simulations[n-1].plot(ax=ax, kind="hist", color= np.random.rand(3,), bins=50)

    ax.set_title(f"dt=1/{N[n-1]}")
    ax.get_legend().remove()
    ax.set_xlabel("Distribution")

plt.show()
#TENDS TO INFINITY AS N TENDS TO INFINITY
```
